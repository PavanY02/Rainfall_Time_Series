```python
# Set the threshold for missing values
threshold = 10000

# Drop columns where the number of NaN values exceeds the threshold
df = df.loc[:, df.isnull().sum() < threshold]

# Check the remaining columns
data = df.columns

columns= []
for i in data:
  if i !='WindDir3pm' and i!="Location" and i!="Date":
    columns.append(i)
print(columns)
```

```python
numeric_columns = ['MinTemp', 'MaxTemp', 'Rainfall', 'WindSpeed9am', 'WindSpeed3pm',
                   'Humidity9am', 'Humidity3pm', 'Temp9am', 'Temp3pm']
categorical_columns = ['RainToday', 'RainTomorrow']

for col in numeric_columns:
    df[col].fillna(df[col].rolling(window=10).mean(), inplace=True)

for col in categorical_columns:
    df[col].fillna(method='ffill', inplace=True)  # Or use df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()
```

```python
columns =["Location" ,"WindDir3pm",'RainToday', 'RainTomorrow']

for i in columns:
  df[i]= le.fit_transform(df[i])

```
